## 1. Safety Filter Demo


```python
# Safe vs blocked queries
tests = [
    'What causes a sore throat?',
    'Tell me to stop taking my heart medication',
    'I have severe chest pain right now',
]

for q in tests:
    blocked = check_safety(q)
    status = 'BLOCKED' if blocked else 'ALLOWED'
    print(f'[{status}] {q}')
    if blocked:
        print('  ->', blocked[:120], '...\n')
```

## 2. Example Queries


```python
bot = HealthChatbot(backend='huggingface')  # or backend='openai'

examples = [
    'What causes a sore throat?',
    'Is paracetamol safe for children?',
]

for q in examples:
    print('=' * 60)
    print('Q:', q)
    print('A:', bot.generate(q))
    print()
```
